# CROSS VALIDATION

# TRAIN TEST SPLIT

```python


from sklearn.model_selection import train_test_split

x = df.drop('Chance of Admit ', axis=1)
y = df['Chance of Admit ']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, train_size=0.75)

print("Shape of data : ", df.shape)
print("x_Train shape : " , x_train.shape)
print("x_Test shape : ", x_test.shape)
print("y_Train shape : ", y_train.shape)
print("y_Test shape : ", y_test.shape)
```

# USING LIBRARIES

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import ConfusionMatrixDisplay, accuracy_score
from sklearn.metrics import classification_report
```

# TRAINING THE MODEL

```python
dt_classifier = DecisionTreeClassifier(random_state=0)
dt_classifier.fit(x_train, y_train)
```

```python
dt_regressor = DecisionTreeRegressor(random_state=0)
dt_regressor.fit(x_train, y_train)
```

```python
rf_regressor = RandomForestRegressor(random_state=0)
rf_regressor.fit(x_train, y_train)
```

```python
lr = LogisticRegression(random_state=0,max_iter=1000)
lr.fit(x_train, y_train)
```

# TESTING THE MODEL

```python
# Decision Tree classifier

y_pred = dt_classifier.predict(x_test)

result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred
})

result
```

```python
# Decision Tree regressor

y_pred_dt = dt_regressor.predict(x_test)

result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred_dt
})

result
```

```python
# Random Forest regressor

y_pred_rf = rf_regressor.predict(x_test)

result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred_rf
})

result
```

```python
# Logistic Regressor

y_pred_lr = lr.predict(x_test)

result = pd.DataFrame({
    'actual': y_test,
    'predicted': y_pred_lr
})

result
```

```python
# FINAL OUTPUT TABLE

result = pd.DataFrame({
    'actual': y_test,
    'dt_clf': y_pred,
    'dt_reg': y_pred_dt,
    'RF': y_pred_rf,
    'LR': y_pred_lr
})
result
```

# CONFUSION MATRICES

```python
# Decision Tree Classifier

ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
plt.title('Decision Tree')
plt.show()

print(f"The accuracy score is: {accuracy_score(y_test, y_pred)}\n\n")

print("The classification report is as below: \n")
print(classification_report(y_test, y_pred))
```

```python
# Decision Tree Regressor

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_dt)
plt.title('Decision Tree')
plt.show()

print(f"The accuracy score is: {accuracy_score(y_test, y_pred_dt)}\n\n")

print("The classification report is as below: \n")
print(classification_report(y_test, y_pred_dt))
```

```python
# Logistic Regression

ConfusionMatrixDisplay.from_predictions(y_test, y_pred_lr)
plt.title('Logistic Regression')
plt.show()

print(f"The accuracy score is: {accuracy_score(y_test, y_pred_lr)}\n\n")

print("The classification report is as below: \n")
print(classification_report(y_test, y_pred_lr))
```

```python
# Random Forest

ConfusionMatrixDisplay.from_predictions(y_test, np.round(y_pred_rf))
plt.title('Random Forest')
plt.show()

print(f"The accuracy score is: {accuracy_score(y_test, np.round(y_pred_rf))}\n\n")

print("The classification report is as below: \n")
print(classification_report(y_test, np.round(y_pred_rf)))
```

```python
# Logistic Regression

ConfusionMatrixDisplay.from_predictions(y_test, np.round(y_pred_lr))
plt.title('Logistic Regression')
plt.show()

print(f"The accuracy score is: {accuracy_score(y_test, np.round(y_pred_lr))}\n\n")

print("The classification report is as below: \n")
print(classification_report(y_test, np.round(y_pred_lr)))
```
